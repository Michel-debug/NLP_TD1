```python
corpora = {}
stats = []

for corpus in tqdm(CORPORA):
    print(corpus)
    texts = []
    for fp in (TEXT_P / corpus).glob("*.txt"):
        with fp.open() as f:
            texts.append(f.read())

    corpora[corpus] = "".join(texts)

    stats.append(
        {
            "corpus": corpus,
            "files_n": len(texts),
            "chars_n": len(corpora[corpus]),
        }
    )

df = pd.DataFrame.from_records(stats, index=["corpus"])
df["text"] = [corpora[corpus] for corpus in corpora]
df
```

### What's going on here? Which text is the richest in characters?

```python
HOP = 100_000
```

```python
#divides general text into batches of len HOP and calculates the number of unique characters

def batch_counts(text, max_len, hop):
    counters = [Counter()]
    for i in range(hop, max_len, hop):
        counters.append(counters[-1] + Counter(text[i - hop : i]))
    return [len(cntr) for cntr in counters]
```

```python
df
```

```python
batch_counters = df.text.map(
    lambda x: batch_counts(x, max_len=df.chars_n.min(), hop=HOP)
)

batch_counters_df = pd.DataFrame(
    dict(zip(batch_counters.index, batch_counters.values))
)
batch_counters_df.index.name = "batch"

batch_counters_df.plot()
plt.show()
```

```python
def avg_len_tokens(tokens):
    return tokens.apply(lambda x : np.mean(list(map(lambda x : len(x),x))))

def max_len_tokens(tokens):
    return tokens.apply(lambda x : max(list(map(lambda x : len(x),x))))

def longest_words(tokens):
    return tokens.apply(lambda corpus_largest_word: corpus_largest_word[list(map(lambda x : len(x),corpus_largest_word)).index(max(list(map(lambda x : len(x),corpus_largest_word))))])
```

```python
print("\n*** average length in the dictionary (on types) ***")
display(avg_len_tokens(tokens))
```

```python
print("\n*** the longest words' length ***")
display(max_len_tokens(tokens))

print("\n*** the longest words ***")
display(longest_words(tokens))
```

```python
def len_freq_counter(text):
    remove_pattern = re.compile(r"[0-9/=+\._-]")
    cleaned_text = [word for word in text if not re.search(remove_pattern, word)]
    counts = Counter([len(word) for word in cleaned_text])
    return list(counts.keys()), list(counts.values())

# Apply len_freq_counter to each category of tokens and store the results in len_freq_df
result = tokens.apply(len_freq_counter)
len_freq_df = pd.DataFrame()
len_freq_df["freq"] = result.apply(lambda x: x[1])
len_freq_df["len"] = result.apply(lambda x: x[0])

# Explode the lists in the 'len' and 'freq' columns into individual rows
len_freq_df = len_freq_df.explode(['freq','len'])

len_freq_df.head(100) 
```

**Explain the graph results:**
\
Zipf's law typically shows that the frequency of any word is inversely proportional to its rank in the frequency table. In simpler terms, the most common word in a text appears about twice as often as the second most common word, three times as often as the third most common, and so on. However, in the provided image, the frequency distribution graph forms an 'L' shape instead of the expected smooth curve descending from the top left to the bottom right. This 'L' shape is primarily due to a few words at the start of the horizontal axis having significantly higher frequencies. Additionally, the graph reflects that a large number of words in the text occur relatively infrequently, contributing to the long tail of the 'L' shape.
